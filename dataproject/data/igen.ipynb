```python
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams.update({"axes.grid":True,"grid.color":"black","grid.alpha":"0.25","grid.linestyle":"--"})
plt.rcParams.update({'font.size': 14})
```

```python
import os 

# Using assert to check that paths exist on computer.
assert os.path.isdir('data/')
assert os.path.isfile('data/bog1.xlsx')


# Print everything in data
os.listdir('data/')
```

```python
filename = 'data/BOG1.xlsx' # open the file and have a look at it
pd.read_excel(filename).head(10)
```

```python
fert = pd.read_excel(filename, skiprows=2)
fert.head(7)
```

```python
fert.drop(columns=[str(i) for i in range(1973, 2010)], inplace=True)
fert.head(5)
```

```python
fert.rename(columns = {'Unnamed: 0':'Age group'}, inplace=True)
fert.head(7)
```

```python
col_dict = {}
for i in range(2010, 2023+1): # range goes from 2008 to but not including 2018
    col_dict[str(i)] = f'fert{i}'

col_dict = {str(i) : f'fert{i}' for i in range(2010,2023+1)}
col_dict
```

```python
fert.rename(columns = col_dict, inplace=True)
fert.head(7)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

fert[['Age group', 'fert2010']].plot(x='Age group', y='fert2010', kind='bar', ax=axes[0, 0])
axes[0, 0].set_xlabel('Age group')
axes[0, 0].set_ylabel('Fertility rate')
axes[0, 0].set_title('Fertility Rate by Age Group (2010)')

fert[['Age group', 'fert2015']].plot(x='Age group', y='fert2015', kind='bar', ax=axes[0, 1])
axes[0, 1].set_xlabel('Age group')
axes[0, 1].set_ylabel('Fertility rate')
axes[0, 1].set_title('Fertility Rate by Age Group (2015)')

fert[['Age group', 'fert2020']].plot(x='Age group', y='fert2020', kind='bar', ax=axes[1, 0])
axes[1, 0].set_xlabel('Age group')
axes[1, 0].set_ylabel('Fertility rate')
axes[1, 0].set_title('Fertility Rate by Age Group (2020)')

fert[['Age group', 'fert2023']].plot(x='Age group', y='fert2023', kind='bar', ax=axes[1, 1])
axes[1, 1].set_xlabel('Age group')
axes[1, 1].set_ylabel('Fertility rate')
axes[1, 1].set_title('Fertility Rate by Age Group (2023)')

plt.tight_layout()
plt.show()
```
